```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker

df = pd.read_csv(r"C:\Users\Jermaine Antonio\Downloads\Data\Car Sales.xlsx - car_data.csv")

#Converts date cp;umn to datetime format
df['Date'] = pd.to_datetime(df['Date']) 
df['Year'] = df['Date'].dt.year 
df['Month'] = df['Date'].dt.month  
df['Day'] = df['Date'].dt.day 

monthly_sales = df.groupby(["Year", "Month"]).size().reset_index(name="Sales Count")



plt.figure(figsize=(8, 6))
sns.boxplot(x="Gender", y="Annual Income", data=df)
plt.title("Customer Income Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Annual Income")

# Force standard number formatting
plt.ticklabel_format(style='plain', axis='y')

# OR Format the Y-axis with commas
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.show()
```
